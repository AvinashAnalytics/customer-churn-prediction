### Linear Regression: Predict MonthlyCharges based on Tenure
Build a simple linear model where the dependent variable is ‘MonthlyCharges’ and the independent variable is ‘tenure’.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming 'churn' is the DataFrame
X = churn[['tenure']]  # Independent variable (Tenure)
y = churn['MonthlyCharges']  # Dependent variable (MonthlyCharges)

# Step 1: Split the dataset into train and test sets (70:30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 2: Build the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 3: Predict the values on the test set
y_pred = model.predict(X_test)

# Step 4: Calculate the error (The difference between actual and predicted values)
error = y_test - y_pred  

# Step 5: Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Print results
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Prediction Error (first few values):\n{error.head()}")
```

 ### Logistic Regression: Predict Churn
 Simple Logistic Regression: Dependent variable = ‘Churn’ & Independent variable = ‘MonthlyCharges’

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Map 'Churn' to 1 (Yes) and 0 (No)
y_churn = churn['Churn'].map({'Yes': 1, 'No': 0})
X_churn = churn[['MonthlyCharges']]  # Independent variable

# Step 1: Split the dataset into train and test sets (65:35 ratio)
X_train_churn, X_test_churn, y_train_churn, y_test_churn = train_test_split(X_churn, y_churn, test_size=0.35, random_state=42)

# Step 2: Build the Logistic Regression model
log_model = LogisticRegression()
log_model.fit(X_train_churn, y_train_churn)

# Step 3: Predict the churn status on the test set
y_pred_churn = log_model.predict(X_test_churn)

# Step 4: Evaluate the model
cm_churn = confusion_matrix(y_test_churn, y_pred_churn)
accuracy_churn = accuracy_score(y_test_churn, y_pred_churn)

# Print results
print(f"Confusion Matrix:\n{cm_churn}")
print(f"Accuracy: {accuracy_churn:.4f}")
```

Multiple Logistic Regression: Dependent variable = ‘Churn’ & Independent variables = ‘tenure’, ‘MonthlyCharges’

```python
# Independent variables: 'tenure' and 'MonthlyCharges'
X_multi_churn = churn[['tenure', 'MonthlyCharges']]  # Independent variables
y_multi_churn = churn['Churn'].map({'Yes': 1, 'No': 0})  # Dependent variable

# Step 1: Split the dataset into train and test sets (80:20 ratio)
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi_churn, y_multi_churn, test_size=0.2, random_state=42)

# Step 2: Build the Logistic Regression model
log_model_multi = LogisticRegression()
log_model_multi.fit(X_train_multi, y_train_multi)

# Step 3: Predict the churn status on the test set
y_pred_multi = log_model_multi.predict(X_test_multi)

# Step 4: Evaluate the model
cm_multi = confusion_matrix(y_test_multi, y_pred_multi)
accuracy_multi = accuracy_score(y_test_multi, y_pred_multi)

# Print results
print(f"Confusion Matrix:\n{cm_multi}")
print(f"Accuracy: {accuracy_multi:.4f}")
```

### Random Forest: Predict Churn using Tenure and MonthlyCharges
a. Random Forest Classifier: Dependent variable = ‘Churn’ & Independent variables = ‘tenure’, ‘MonthlyCharges’

```python
from sklearn.ensemble import RandomForestClassifier

# Independent variables: 'tenure' and 'MonthlyCharges'
X_rf = churn[['tenure', 'MonthlyCharges']]  # Independent variables
y_rf = churn['Churn'].map({'Yes': 1, 'No': 0})  # Dependent variable

# Step 1: Split the dataset into train and test sets (70:30 ratio)
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X_rf, y_rf, test_size=0.3, random_state=42)

# Step 2: Build the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Step 3: Predict the churn status on the test set
y_pred_rf = rf_model.predict(X_test_rf)

# Step 4: Evaluate the model
cm_rf = confusion_matrix(y_test_rf, y_pred_rf)
accuracy_rf = accuracy_score(y_test_rf, y_pred_rf)

# Print results
print(f"Confusion Matrix:\n{cm_rf}")
print(f"Accuracy: {accuracy_rf:.4f}")
```
